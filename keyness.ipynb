```python
# import packages for preparing the text for bigram analysis
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.util import bigrams
import json

big_corpus = pd.read_table('final_result/big_corpus_result.tsv')
```

```python
big_corpus = pd.read_table('final_result/big_corpus_result.tsv')
```

```python
processed_text = [json.loads(text.replace('\'', '"')) for text in big_corpus['processed_text']]

def a():
    global all_bigrams, negative_bigrams, neutral_bigrams, positive_bigrams
    all_bigrams = []
    for t in processed_text:
        for bigram in list(bigrams(t)):
            all_bigrams.append(bigram[0] + '_' + bigram[1])

    negative_bigrams = []
    neutral_bigrams = []
    positive_bigrams = []

    for i, score in enumerate(big_corpus['sent'].tolist()):
        if score == 0:
            negative_bigrams.append(all_bigrams[i])
        if score == 1:
            neutral_bigrams.append(all_bigrams[i])
        if score == 2:
            positive_bigrams.append(all_bigrams[i])

def b():
    global all_bigrams, negative_bigrams, neutral_bigrams, positive_bigrams
    all_bigrams = [[bigram[0] + '_' + bigram[1] for bigram in list(bigrams(t))] for t in processed_text]

    negative_bigrams = []
    neutral_bigrams = []
    positive_bigrams = []

    for i, score in enumerate(big_corpus['sent'].tolist()):
        if score == 0:
            negative_bigrams.append(all_bigrams[i])
        if score == 1:
            neutral_bigrams.append(all_bigrams[i])
        if score == 2:
            positive_bigrams.append(all_bigrams[i])
```

```python
from corpus_toolkit import corpus_tools as ct
```

```python
corp1freq = ct.frequency(ct.tokenize(positive_bigrams))
corp2freq = ct.frequency(ct.tokenize(all_bigrams))
corp_key = ct.keyness(corp1freq,corp2freq, effect = "%diff")
ct.head(corp_key, hits = 100) #to display top hits
```

```python
b()
from keyness import log_likelihood

negative_keys = log_likelihood(negative_bigrams, all_bigrams , dummy_zero=0)[:500]
positive_keys = log_likelihood(positive_bigrams, all_bigrams , dummy_zero=0)[:500]
neutral_keys = log_likelihood(neutral_bigrams, all_bigrams , dummy_zero=0)[:500]

```

```python
negative_keys
neg = pd.DataFrame(negative_keys, columns =['Bigram', 'Loglikelihood', 'Frequency', 'Frequency in Sent'])
pos = pd.DataFrame(positive_keys, columns =['Bigram', 'Loglikelihood', 'Frequency', 'Frequency in Sent'])
neu = pd.DataFrame(neutral_keys, columns =['Bigram', 'Loglikelihood', 'Frequency', 'Frequency in Sent'])
```

```python
def create_key(data):
    freavg = data['Frequency'].mean()
    avg2 = data['Frequency in Sent'].mean()
    data['keyness'] = data['Loglikelihood'] * data['Frequency'] / data['Frequency in Sent'] 
    return data
neg = create_key(neg)
pos = create_key(pos)
neu = create_key(neu)
```
